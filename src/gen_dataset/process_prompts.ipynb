```python
import pandas as pd
from models import BADStatement, StatementMetadata
import pickle
from pathlib import Path
```

```python
def create_bads_statements(stmt: StatementMetadata) -> BADStatement:
    return BADStatement(
        type=stmt.statement.type,
        marked_text=stmt.statement.marked_text,
        themes=stmt.themes,
        target_groups=stmt.target_groups,
        data_source=stmt.data_source,
        is_harmful=stmt.is_harmful
    )
```

```python
print(Path.cwd())
outputs_dir = Path.cwd() / "outputs"
with open(outputs_dir / "structured_responses_other.pkl", "rb") as f:
    stored_pickle = pickle.load(f)
print(len(stored_pickle))
stored_pickle[:3]
```

```python
final_statements = map(create_bads_statements, stored_pickle)
final_statements = list(final_statements)
print(len(final_statements))
final_statements[:5]
```

```python
df = pd.DataFrame([stmt.__dict__ for stmt in final_statements])

df.to_csv("final_statements.csv", mode='a', header=False, index=False)
df.head(5)
```
